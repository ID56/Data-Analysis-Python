<h1>Bivariate Plotting</h1>

Here we'll explore various types of bivariate plots, using the pokemon dataset.

```python
import pandas as pd
```

```python
pokemon=pd.read_csv('pokemon.csv')
pokemon.head()
```

<strong>Scatter Plots: </strong>Plots dots of a mapping between two data columns.

```python
pokemon.plot.scatter(x='attack', y='defense')
```

<strong>Hex plots: </strong>In scatter plots, it is difficult to understand the frequency of values when many values overlap on the same grid position. Hex bins display frequency with a color scale. 

```python
pokemon.plot.hexbin(x='attack',y='speed',gridsize=15)
```

<strong> Stacked Bar Plots: </strong> 


```python
pokemon_stats_legendary = pokemon.groupby(['is_legendary']).mean()[['attack','defense']]
pokemon_stats_legendary
```

```python
pokemon_stats_legendary.plot.bar(stacked=False)
```

```python
pokemon_stats_by_generation = pokemon.groupby('generation').mean()[['hp', 'attack', 'defense', 'sp_attack', 'sp_defense', 'speed']]
pokemon_stats_by_generation
```

```python
pokemon_stats_by_generation.plot.bar(stacked=True)
```

<strong>Line chart: </strong>

```python
pokemon_stats_by_generation.plot.line()
```
